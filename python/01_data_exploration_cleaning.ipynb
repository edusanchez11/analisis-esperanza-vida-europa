```python
# carga librerias necesarias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# carga datasets

health_care_expenditure = pd.read_csv("C:\\Users\\eduar\\OneDrive\\Escritorio\\Master_DataScience_AI\\Proyecto_Comparativa_acceso_y_gasto_en_salud_entre_países_europeos\\data\\health_care_expenditure.csv")
life_expectancy = pd.read_csv("C:\\Users\\eduar\\OneDrive\\Escritorio\\Master_DataScience_AI\\Proyecto_Comparativa_acceso_y_gasto_en_salud_entre_países_europeos\\data\\life_expectancy.csv")
poblation = pd.read_csv(r"C:\Users\eduar\OneDrive\Escritorio\Master_DataScience_AI\Proyecto_Comparativa_acceso_y_gasto_en_salud_entre_países_europeos\data\poblation.csv")
```

```python
# exploracion inicial dataset

health_care_expenditure.head()
```

```python
# renombrar columnas
health_care_expenditure = health_care_expenditure.rename(columns={
    'geo': 'country',
    'TIME_PERIOD': 'year',
    'OBS_VALUE': 'expenditure',
    'unit': 'uds',
    'icha11_hf': 'financing schemes'
})

#eliminar innecesarias

health_care_expenditure = health_care_expenditure[['country', 'year', 'expenditure', 'financing schemes','uds']]

health_care_expenditure.head()
```

```python
health_care_expenditure.info()
```

```python
# eliminar filas vacias

health_care_expenditure.isnull().sum()

health_care_expenditure = health_care_expenditure.dropna()
```

```python
# valores de gasto total en salud

health_care_expenditure = health_care_expenditure[health_care_expenditure['financing schemes'] == 'All financing schemes']
```

```python
health_care_expenditure['expenditure'].describe()
```

```python
health_care_expenditure['expenditure'] = health_care_expenditure['expenditure'] * 1000000

health_care_expenditure['uds'] = '€'
```

```python
# Ver todos los valores únicos de age

print("Valores únicos en 'age':")
print(life_expectancy['age'].unique())

# Ver cuántos registros hay por cada valor de age
print("\nDistribución de valores:")
print(life_expectancy['age'].value_counts())
```

```python
# esperanza de vida al nacer (la más común)

life_expectancy = life_expectancy[life_expectancy['age'] == 'Less than 1 year']
```

```python
# renombrar columnas
life_expectancy = life_expectancy.rename(columns={
    'geo': 'country',
    'TIME_PERIOD': 'year',
    'OBS_VALUE': 'expectancy',
    'unit': 'uds'
})

life_expectancy = life_expectancy[['country', 'year', 'expectancy', 'uds', 'sex']]

life_expectancy.head(20)
```

```python
life_expectancy.isnull().sum()

life_expectancy.dropna()
```

```python
life_expectancy['expectancy'].describe()
```

```python
life_expectancy_filtered = life_expectancy[life_expectancy['sex'].isin(['Females', 'Males'])]

life_expectancy_filtered.head()

# print(f"Filas originales: {len(life_expectancy)}")
# print(f"Filas después del filtro: {len(life_expectancy_total)}")
# print(f"Valores únicos en 'sex': {life_expectancy_total['sex'].unique()}")
```

```python
poblation.head()

```

```python
# renombrar columnas
poblation = poblation.rename(columns={
    'geo': 'country',
    'TIME_PERIOD': 'year',
    'sex': 'sex',
    'OBS_VALUE': 'people',
})

poblation = poblation[['country', 'year','sex', 'people']]

poblation.head(20)
```

```python
# unir todos los datasets

df = health_care_expenditure.merge(life_expectancy_filtered, on = ['country', 'year'], how = 'inner')
df = df.merge(poblation, on=['country', 'year', 'sex'], how='inner')


# renombrar columnas

df = df.rename(columns={
    'uds_x': 'expenditure_unit',
    'uds_y': 'life_expectancy_unit'
})

df.head(20)
```

```python
# Verificar datos faltantes
print("Valores nulos por columna:")
print(df.isnull().sum())

# Información general
print(f"\nPaíses únicos: {df['country'].nunique()}")
print(f"Años: {df['year'].min()} - {df['year'].max()}")
print(f"Total registros: {len(df)}")
```

```python
# Ver todas las columnas disponibles
print("Columnas del DataFrame:")
print(df.columns.tolist())

# O ver información completa del DataFrame
print(df.info())
```

```python
# Eliminar filas donde el país sea 'European Union' o 'Euro Area'
df = df[~df['country'].isin(['European Union', 'Euro Area', 'European Union - 27 countries (from 2020)', 'Euro area - 19 countries  (2015-2022)', 'Euro area – 20 countries (from 2023)'])]
```
